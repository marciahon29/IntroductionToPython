```python
# First it is very important to load the library
import matplotlib.pyplot as plt #.pyplot as plt
import numpy as np
```

# Simple Plot

```python
# A very simple plot - blue line
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.ylabel('y-axis numbers')
plt.xlabel('x-axis numbers')
plt.show()
```

```python
# A very simple plot - red circles
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')
plt.axis([0, 6, 0, 20])
plt.ylabel('y-axis numbers')
plt.xlabel('x-axis numbers')
plt.show()
```

```python
# Another Simple example
t = np.arange(0., 5., 0.2) # 0 to 5 at 0.2 interval

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

# Categorical Values

```python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(figsize=(9, 3)) # width x height in inches

plt.subplot(131) # 1 row, 3 columns, position 1
plt.bar(names, values)
plt.subplot(132)
plt.scatter(names, values)
plt.subplot(133)
plt.plot(names, values)
plt.suptitle('Categorical Plotting')
plt.show()
```

# Histogram

```python
import numpy as np
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000) # Normal distribution of 10000 variables

# the histogram of the data
plt.hist(x, bins=50, density=1, facecolor='g', alpha=0.75) # Histogram: bins=number of equal size x-axis, density=1:value of integral is equal to 1; alpha=transparency

plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```

# Multiple Figures and Axes

```python
# Custom Function
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1) # Range from 0 to 5 with 0.1 apart
t2 = np.arange(0.0, 5.0, 0.02) # Range from 0 to 5 with 0.02 p

plt.figure()
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k') # 'k' = black

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

# Annotation

```python
ax = plt.subplot(111)

t = np.arange(0.0, 5.0, 0.01) # Range 0 to 5 at 0.01 apart
s = np.cos(2*np.pi*t)
line = plt.plot(t, s, lw=5) #lw: line width

plt.annotate('local max', xy=(2, 1), xytext=(3, 1.5), # Location of arrow and Location of text
             arrowprops=dict(facecolor='black', shrink=0.05),
             )

plt.ylim(-2, 2)
plt.show()
```

# Logarithmic and other nonlinear axes

```python
# Fixing random state for reproducibility
np.random.seed(19680801)

# make up some data in the open interval (0, 1)
y = np.random.normal(loc=0.5, scale=0.4, size=1000) #Generates sample numbers from Gaussian, loc=center, scale=standard deviation, size=number of values
y = y[(y > 0) & (y < 1)] # Make sure y is in the range of 0 to 1
y.sort()
x = np.arange(len(y)) #integers with range of 0 to len(y) separated by 1

# plot with various axes scales
plt.figure()

# linear
plt.subplot(221)
plt.plot(x, y)
plt.yscale('linear') # Linear scale
plt.title('linear')
plt.grid(True)

# log
plt.subplot(222)
plt.plot(x, y)
plt.yscale('log') # Log scale
plt.title('log')
plt.grid(True)

# symmetric log
plt.subplot(223)
plt.plot(x, y - y.mean())
# Since the values close to zero tend toward infinity, there is a need to have a range around zero that is linear.
# The parameter linthresh allows the user to specify the size of this range (-linthresh, linthresh).
plt.yscale('symlog', linthresh=0.01) # symmetric log
plt.title('symlog')
plt.grid(True)

# logit
plt.subplot(224)
plt.plot(x, y)
plt.yscale('logit') # logit
plt.title('logit')
plt.grid(True)

# Adjust the subplot layout, because the logit one may take more space
# than usual, due to y-tick labels like "1 - 10^{-3}"
plt.subplots_adjust(top=1.50, bottom=0.00, left=0.10, right=0.95, hspace=0.35,
                    wspace=0.35)
    # left = 0.125, the left side of the subplots of the figure
    # right = 0.9, the right side of the subplots of the figure
    # bottom = 0.1, the bottom of the subplots of the figure
    # top = 0.9, the top of the subplots of the figure
    # wspace = 0.2, the amount of width reserved for space between subplots, expressed as a fraction of the average axis width
    # hspace = 0.2, the amount of height reserved for space between subplots, expressed as a fraction of the average axis height

plt.show()
```

# HOMEWORK
Plot 3 sin waves:
    
sin(x)
sin(2*x)
sin(3*x)

Use 100 digits
Include a legend

To generate the x, you need to use the following:
x = np.linspace(-np.pi,np.pi,100)

What this does is get 100 equally spaced values of radians from -pi to +pi

```python
#SOLUTION
import matplotlib.pyplot as plt
import numpy as np

# 100 linearly spaced numbers
x = np.linspace(-np.pi,np.pi,100)

p = np.sin(x) # y = sin(x)
q = np.sin(2*x) # y = sin(2x)
r = np.sin(3*x) # y = sin(3x)

# plot the functions, with labels
plt.plot(x,p, 'b-', label='y=sin(x)')
plt.plot(x,q, 'c-', label='y=sin(2x)')
plt.plot(x,r, 'm-', label='y=sin(3x)')

plt.legend(loc='upper left')

# show the plot
plt.show()
```
